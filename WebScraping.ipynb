```python
url = 'https://en.wikipedia.org/wiki/List_of_Indonesian_provinces_by_GDP'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')[1]
```

```python
table = soup.find_all('table')[1]
```

```python
indoprovgdp_row_title = table.find_all('th')
```

```python
row_title = [title.text.strip() for title in indoprovgdp_row_title]
del row_title[6:]
print(row_title)
```

```python
datatitle = pd.DataFrame(columns = row_title)
datatitle
```

```python
indoprovgdp_body_table = table.find_all('tr')
for row in indoprovgdp_body_table[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(datatitle)
    datatitle.loc[length] = individual_row_data
```

```python
df = datatitle[datatitle['Rank'] != '-']
```

```python
df = df.drop('Rank', axis=1)
```
